```python
import os
import sys
import collections
import scipy.spatial.distance
import scipy.stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.pylabtools import figsize
sns.set(font_scale=2)
sns.set_style("white")
%matplotlib inline
```

```python
polygamy_data = pd.read_excel('PwoP vs SA Summary.xlsx', sheetname = 0)
monogamy_data = pd.read_excel('PwoP vs SA Summary.xlsx', sheetname = 1)
```

```python
current_palette = sns.color_palette('dark')
sns.palplot(current_palette)
```

# Compare mating systems

```python
merged = pd.merge(polygamy_data, monogamy_data, on=u'dir_name ', suffixes = ('poly', 'mono'))
```

```python
sns.jointplot(x = u' PWOP_Ne poly',  y = u' PWOP_Ne mono', data = merged, color = current_palette[0], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r"))
plt.xlim(0, 700)
plt.ylim(0,700)
plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('PWOP', loc = 'left', y=.9, x=.1)
plt.show()

sns.jointplot(x = u' Ne-Nonrand poly',  y = u' Ne-Nonrand mono', data = merged, color = current_palette[1], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r"))
plt.xlim(0, 700)
plt.ylim(0,700)
plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Colony', loc = 'left', y=.9, x=.1)
plt.show()
```

# Compare estimation methods

```python
polygamy_data['mating'] = 'polygamy'
monogamy_data['mating'] = 'monogamy'
data = pd.concat([polygamy_data, monogamy_data])
```

```python
sns.jointplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = data, color = current_palette[2], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r", loc = 'lower right'), joint_kws=dict(fit_reg = False), scatter_kws=dict(s=40,alpha=1))
plt.xlim(0,700)
plt.ylim(0,700)
plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Both mating systems', loc = 'left', y=.9, x=.1)
plt.show()
```

```python
sns.jointplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = polygamy_data, color = current_palette[3], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r", loc = 'lower right'), joint_kws=dict(fit_reg = False), scatter_kws=dict(s=40,alpha=1))
plt.xlim(0,700)
plt.ylim(0,700)
plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Ploygamy', loc = 'left', y=.9, x=.1)
plt.show()

sns.jointplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = monogamy_data, color = current_palette[4], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r", loc = 'lower right'), joint_kws=dict(fit_reg = False), scatter_kws=dict(s=40,alpha=1))
plt.xlim(0,700)
plt.ylim(0,700)
plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Monogamy', loc = 'left', y=.9, x=.1)
plt.show()
```

```python
sns.jointplot(x = u'Differencepoly', y = u'Differencemono', data = merged, color = current_palette[5], kind = 'reg',
              stat_func=scipy.stats.pearsonr, size=8, ratio=5, marginal_kws=dict(bins = 7,kde = False),
                 annot_kws=dict(stat="r", loc = 'lower right'), joint_kws=dict(fit_reg = True, x_jitter = .1, y_jitter = .1), scatter_kws=dict(s=40,alpha=1))
plt.xlim(-10,5)
plt.ylim(-10,10)
plt.plot([-10,10], [-10,10], 'k--', alpha = .5)
plt.title('Difference', loc = 'left', y=.9, x=.1)
plt.show()
```

```python
sns.residplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = data, color = current_palette[5] )

#plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Residuals - both mating systems', loc = 'left', y=.9, x=.02)
plt.show()
```

```python
sns.residplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = monogamy_data, color = current_palette[5] )

#plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Residuals - mongamy', loc = 'left', y=.9, x=.02)
plt.show()

sns.residplot(x = u' Ne-Nonrand ', y = u' PWOP_Ne ', data = polygamy_data, color = current_palette[5] )

#plt.plot([0, 700], [0, 700], 'k--', alpha = .5)
plt.title('Residuals - polygamy', loc = 'left', y=.9, x=.02)
plt.show()
```
